## Area ID

```python
P_A  = cite_fake_new["Paper Area"].apply(lambda x: x.split("; "))
P_A
```

```python
paper_area = []

for item in P_A:
    paper_area += item

paper_area
```

```python
paper_area = list(set(paper_area))
```

```python
paper_area.sort()
paper_area
```

```python
db = 0
for a in paper_area:
    if a == "Natural language processing":
        db+=1
        
print(db)
        
```

```python
pd.DataFrame(paper_area).rename(columns = {0:"Area"}).to_csv("./Processed_data/Area_ID.csv",index= True, 
                                                             encoding="utf_8_sig",
                                                            index_label= "Area_ID")
```
