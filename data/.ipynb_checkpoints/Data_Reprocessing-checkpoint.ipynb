```python
def change_area(data):

    add_area = ["Databases","Machine learning","Natural language processing","Nothing"]
    add_keywords = np.random.choice(add_area, p = [0.3,0.15,0.1,0.45])
    if pd.isnull(data) == True:
        return np.random.choice(add_area[0:3], p = [0.5,0.3,0.2])
    else:
        if add_keywords != "Nothing" and add_keywords not in data:
            
            return (data + ("; " +add_keywords))
        else:
            return data
```

```python
cite_fake["Paper Area"] = cite_fake["Paper Area"].apply(lambda x: change_area(x))
```

```python
word_count = dict()

for conf in cite_fake["Source title"].unique():
    d = dict()
    df = cite_fake.loc[cite_fake["Source title"] == conf]["Paper Area"]
    
    for i,keywords in enumerate(df):
        PA = keywords.split("; ")
        for keyword in PA:
            if keyword not in d.keys():
                d[keyword] = 1
            else:
                d[keyword] +=1
                
    word_count[conf] = d
    
    
```

```python
word_count.keys()
```

```python
from collections import Counter
```

```python
a = []
for key,value in word_count.items():
    c = Counter(value)
    most_common = c.most_common(6)
    k = [i[0] for i in most_common ]
    a.append(k)
    print(most_common)
    print(k)
    
```

```python
a[3].remove('Databases')
a[8].remove('Databases')
```
