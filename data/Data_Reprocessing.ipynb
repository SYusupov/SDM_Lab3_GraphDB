## Data Preprocessing

```python
review = review.drop(["Source title","Year","Type","Edition","Authors_list","Title"],axis = 1)
review
```

```python
approval_list = np.zeros((890,))
for row_id in range(review.shape[0]):
    if review.iloc[row_id].Approval == "Yes":
        approval_list[review.iloc[row_id].TitleID] = 1
    else:
        approval_list[review.iloc[row_id].TitleID] = 0
        

approval_list2 = list(map(lambda x: "Yes" if x==1 else "No",approval_list))
approval_list2
```

```python
cite_fake["Approval"] = approval_list2
```

```python
cite_fake
```

```python
cite_fake["Volume"] = cite_fake["Volume"].apply(lambda x: np.random.randint(low=1,high=200) if pd.isnull(x) == True else x)
cite_fake
```

```python
for row in range(cite_fake.shape[0]):
    if cite_fake["Approval"].iloc[row] == "No":
        cite_fake["Volume"].iloc[row] = np.nan
cite_fake
```

```python
cite_fake["Proceeding"] = np.where(cite_fake["Type"] == "Conference_Workshop",cite_fake["Volume"],np.nan)
```

```python
cite_fake["Volume"] = np.where(cite_fake["Type"] == "Journal",cite_fake["Volume"],np.nan)
```

```python
cite_fake
```

```python
np.unique(cite_fake.loc[cite_fake["Type"] == "Conference_Workshop"]["Source title"])
```

```python
conference_type = []
for i in range(cite_fake.shape[0]):
    if cite_fake.iloc[i].Type == "Conference_Workshop":
        if cite_fake.iloc[i]["Source title"] == 'ASEE Annual Conference and Exposition, Conference Proceedings':
            conference_type.append("Workshop")
        elif cite_fake.iloc[i]["Source title"] == 'Optics InfoBase Conference Papers':
            conference_type.append("Symposium")
        elif cite_fake.iloc[i]["Source title"] == 'Proceedings of SPIE - The International Society for Optical Engineering':
            conference_type.append("Expert Group")
        elif cite_fake.iloc[i]["Source title"] == 'Proceedings of the ASME Design Engineering Technical Conference':
            conference_type.append("Regular Conference")
        else:
            conference_type.append("Regular Conference")
    else:
        conference_type.append(np.nan)
conference_type
```

```python
cite_fake["Conference Type"] = conference_type
cite_fake
```

```python
paper_type = []
for i in range(cite_fake.shape[0]):
    if cite_fake.iloc[i].Type == "Conference_Workshop":
        paper_type.append(np.random.choice(["Full Paper","Short Paper","Demo Paper","Poster"]))
    else:
        paper_type.append(np.nan)
        
paper_type
        
```

```python
np.random.choice(["Full Paper","Short Paper","Demo Paper","Poster"])
```

```python
cite_fake["Paper Type"] = paper_type
```

```python
cite_fake
```

```python
cite_fake["Type"] = cite_fake["Type"].apply(lambda x: "Conference" if x== "Conference_Workshop" else x)
cite_fake
```

```python
cite_fake.to_csv("./Processed_data/Cite_fake_New.csv",index=False,encoding="utf_8_sig")
```
